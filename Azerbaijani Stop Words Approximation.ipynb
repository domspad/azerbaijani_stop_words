```python
from __future__ import print_function, division
from codecs import open
import re
from collections import Counter
%matplotlib inline
import matplotlib.pyplot as plt

```

```python
import json
with open('azerarticles.json','r','utf-8') as f : 
    texts = json.load(f)
```

```python
alltext = '\n'.join(t['text'] for t in texts)
```

```python
print(alltext[:400])
```

```python
words = re.findall(ur'\w+',alltext,re.UNICODE)
from collections import Counter


freqs = Counter(words)
freqs.most_common(20)
N = 160
cwords, c = zip(*freqs.most_common(N))

```

```python
plt.figure()
plt.bar(range(N),c)
# plt.xticks(range(N),cwords)
plt.show()
ncwords = [c for c in cwords if not any(w.isdigit() for w in c)
                              if not c.istitle()]
print(len(ncwords))
print(u'\n'.join(ncwords))
```
